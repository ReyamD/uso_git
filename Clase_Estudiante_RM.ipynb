```python
class Empleado:
    """Clase para representar un empleado con sus atributos y validaciones"""

    def __init__(self, nombre, apellido, cedula, email, semestre, materias):
        """Constructor de la clase Empleado"""
        self.__nombre = nombre
        self.__apellido = apellido
        self.__cedula = cedula
        self.__email = email
        self.__semestre = semestre
        self.__materias = materias

    # Propiedad y setter para NOMBRE
    @property
    def nombre(self):
        return self.__nombre

    @nombre.setter
    def nombre(self, nombre):
        if nombre and isinstance(nombre, str) and nombre.strip():
            self.__nombre = nombre.strip()
        else:
            raise ValueError("El nombre no puede estar vacío")

    # Propiedad y setter para APELLIDO
    @property
    def apellido(self):
        return self.__apellido

    @apellido.setter
    def apellido(self, apellido):
        if apellido and isinstance(apellido, str) and apellido.strip():
            self.__apellido = apellido.strip()
        else:
            raise ValueError("El apellido no puede estar vacío")

    # Propiedad y setter para CÉDULA (validación: 10 dígitos numéricos)
    @property
    def cedula(self):
        return self.__cedula

    @cedula.setter
    def cedula(self, cedula):
        # Convertir a string y validar que sean exactamente 10 dígitos
        cedula_str = str(cedula)
        if cedula_str.isdigit() and len(cedula_str) == 10:
            self.__cedula = cedula_str
        else:
            raise ValueError("La cédula debe contener exactamente 10 dígitos numéricos")

    # Propiedad y setter para EMAIL
    @property
    def email(self):
        return self.__email

    @email.setter
    def email(self, email):
        # Validación básica de email
        if email and isinstance(email, str) and '@' in email and '.' in email:
            self.__email = email.strip()
        else:
            raise ValueError("El email no es válido. Debe contener '@' y un dominio válido")

    # Propiedad y setter para SEMESTRE (validación: 1, 2, 3, 4, 5, 6, 7, 8, 9)
    @property
    def semestre(self):
        return self.__semestre

    @semestre.setter
    def semestre(self, semestre):
        try:
            semestre_int = int(semestre)
            # Validación: semestre debe ser 1, 2, 3, 4, 5, 6, 7, 8 o 9
            if semestre_int in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
                self.__semestre = semestre_int
            else:
                raise ValueError("El semestre debe ser un valor entre 1 y 9 (1, 2, 3, 4, 5, 6, 7, 8, 9)")
        except ValueError:
            raise ValueError("El semestre debe ser un número entero entre 1 y 9")

    # Propiedad y setter para MATERIAS
    @property
    def materias(self):
        return self.__materias

    @materias.setter
    def materias(self, materias):
        if isinstance(materias, list) and len(materias) > 0:
            self.__materias = materias
        else:
            raise ValueError("Las materias deben ser una lista no vacía")

    # Propiedad adicional para obtener nombre completo
    @property
    def nombre_completo(self):
        """Retorna el nombre completo del empleado"""
        return f"{self.__nombre} {self.__apellido}"

    # Métodos adicionales
    def agregar_materia(self, materia):
        """Agrega una materia a la lista del empleado"""
        if materia and isinstance(materia, str):
            self.__materias.append(materia)
            return f"Materia '{materia}' agregada correctamente"
        else:
            raise ValueError("La materia debe ser una cadena de texto válida")

    def eliminar_materia(self, materia):
        """Elimina una materia de la lista del empleado"""
        if materia in self.__materias:
            self.__materias.remove(materia)
            return f"Materia '{materia}' eliminada correctamente"
        else:
            return f"La materia '{materia}' no se encuentra en la lista"

    def mostrar_materias(self):
        """Muestra todas las materias del empleado"""
        if self.__materias:
            return ", ".join(self.__materias)
        else:
            return "No tiene materias inscritas"

    def obtener_informacion(self):
        """Método para obtener información completa del empleado"""
        return (f"Nombre completo: {self.nombre_completo}\n"
                f"Cédula: {self.__cedula}\n"
                f"Email: {self.__email}\n"
                f"Semestre: {self.__semestre}\n"
                f"Materias: {self.mostrar_materias()}")

    def __str__(self):
        """Método especial que retorna un diccionario con todos los atributos de la instancia"""
        return str({
            'nombre': self.__nombre,
            'apellido': self.__apellido,
            'cedula': self.__cedula,
            'email': self.__email,
            'semestre': self.__semestre,
            'materias': self.__materias
        })


# Ejemplo de uso
if __name__ == "__main__":
    print("=== DEMOSTRACIÓN DE LA CLASE EMPLEADO ===\n")
    print("SEMESTRES VÁLIDOS: 1, 2, 3, 4, 5, 6, 7, 8, 9\n")

    # Crear empleados con diferentes semestres válidos
    try:
        empleado1 = Empleado(
            nombre="Ana",
            apellido="García",
            cedula="1234567890",
            email="ana.garcia@empresa.com",
            semestre=3,  # Semestre válido
            materias=["Python", "Base de Datos", "React"]
        )

        empleado2 = Empleado(
            nombre="Carlos",
            apellido="López",
            cedula="0987654321",
            email="carlos.lopez@empresa.com",
            semestre=5,  # Semestre válido
            materias=["JavaScript", "Node.js", "MongoDB"]
        )

        empleado3 = Empleado(
            nombre="María",
            apellido="Rodríguez",
            cedula="1112223334",
            email="maria.rodriguez@empresa.com",
            semestre=9,  # Semestre válido (máximo)
            materias=["Java", "Spring Boot", "MySQL"]
        )

        empleado4 = Empleado(
            nombre="Pedro",
            apellido="Martínez",
            cedula="5556667778",
            email="pedro.martinez@empresa.com",
            semestre=1,  # Semestre válido (mínimo)
            materias=["Introducción a la Programación", "Matemáticas Básicas"]
        )

        print("=== EMPLEADOS CREADOS EXITOSAMENTE ===\n")

        # Mostrar información de los empleados
        for empleado in [empleado1, empleado2, empleado3, empleado4]:
            print(f"Empleado: {empleado.nombre_completo}")
            print(f"  Cédula: {empleado.cedula}")
            print(f"  Email: {empleado.email}")
            print(f"  Semestre: {empleado.semestre}")
            print(f"  Materias: {empleado.mostrar_materias()}")
            print()

        # Modificar semestre con valores válidos
        print("=== MODIFICANDO SEMESTRES CON VALORES VÁLIDOS ===")
        print(f"Semestre actual de Ana: {empleado1.semestre}")
        empleado1.semestre = 7  # Cambiar a semestre 7 (válido)
        print(f"Semestre modificado: {empleado1.semestre}")

        empleado1.semestre = 1  # Cambiar a semestre 1 (válido)
        print(f"Semestre modificado a 1: {empleado1.semestre}")

        empleado1.semestre = 9  # Cambiar a semestre 9 (válido)
        print(f"Semestre modificado a 9: {empleado1.semestre}")
        print()

        # Usar el método __str__
        print("=== USANDO __str__ ===")
        print(empleado1)
        print()

    except ValueError as e:
        print(f"Error: {e}")

    # Pruebas de validación de semestres (valores inválidos)
    print("\n" + "="*60)
    print("\n=== PRUEBAS DE VALIDACIÓN DE SEMESTRES ===")
    print("SOLO SE ACEPTAN VALORES: 1, 2, 3, 4, 5, 6, 7, 8, 9\n")

    # Prueba 1: Semestre = 0 (inválido)
    print("Prueba 1: Intentar crear empleado con semestre = 0")
    try:
        empleado_invalido = Empleado(
            nombre="Prueba",
            apellido="Uno",
            cedula="1234567890",
            email="test@email.com",
            semestre=0,  # Inválido
            materias=["Prueba"]
        )
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 2: Semestre = 10 (inválido)
    print("Prueba 2: Intentar crear empleado con semestre = 10")
    try:
        empleado_invalido = Empleado(
            nombre="Prueba",
            apellido="Dos",
            cedula="0987654321",
            email="test@email.com",
            semestre=10,  # Inválido
            materias=["Prueba"]
        )
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 3: Semestre negativo (inválido)
    print("Prueba 3: Intentar crear empleado con semestre = -5")
    try:
        empleado_invalido = Empleado(
            nombre="Prueba",
            apellido="Tres",
            cedula="1112223334",
            email="test@email.com",
            semestre=-5,  # Inválido
            materias=["Prueba"]
        )
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 4: Semestre decimal (inválido)
    print("Prueba 4: Intentar crear empleado con semestre = 3.5")
    try:
        empleado_invalido = Empleado(
            nombre="Prueba",
            apellido="Cuatro",
            cedula="5556667778",
            email="test@email.com",
            semestre=3.5,  # Inválido
            materias=["Prueba"]
        )
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 5: Semestre como string no numérico (inválido)
    print("Prueba 5: Intentar crear empleado con semestre = 'cinco'")
    try:
        empleado_invalido = Empleado(
            nombre="Prueba",
            apellido="Cinco",
            cedula="9998887776",
            email="test@email.com",
            semestre="cinco",  # Inválido
            materias=["Prueba"]
        )
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 6: Modificar semestre existente a valor inválido
    print("Prueba 6: Modificar semestre de empleado existente a 15")
    try:
        empleado_valido = Empleado(
            nombre="Válido",
            apellido="Test",
            cedula="1231231234",
            email="valido@email.com",
            semestre=5,  # Válido
            materias=["Test"]
        )
        print(f"  Semestre actual: {empleado_valido.semestre}")
        empleado_valido.semestre = 15  # Inválido
    except ValueError as e:
        print(f"  ❌ Error capturado: {e}\n")

    # Prueba 7: Mostrar valores válidos
    print("=== SEMESTRES VÁLIDOS ===")
    print("✅ Semestres aceptados: 1, 2, 3, 4, 5, 6, 7, 8, 9")

    print("\n=== EJEMPLOS DE SEMESTRES VÁLIDOS ===")
    for semestre_valido in [1, 2, 3, 4, 5, 6, 7, 8, 9]:
        try:
            empleado_ejemplo = Empleado(
                nombre="Ejemplo",
                apellido=f"Semestre{semestre_valido}",
                cedula=f"123456789{semestre_valido}",
                email=f"ejemplo{semestre_valido}@email.com",
                semestre=semestre_valido,
                materias=["Ejemplo"]
            )
            print(f"  ✅ Semestre {semestre_valido} - Válido")
        except ValueError:
            print(f"  ❌ Semestre {semestre_valido} - Inválido")
```
